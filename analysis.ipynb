# Mathematical Explanation: Single Supplier vs Multiple Suppliers vs No REC

This document provides a comprehensive mathematical analysis of how REC (Renewable Energy Community) sharing affects imbalances differently in single-supplier vs multiple-supplier scenarios.

## Notation

Let:

- $L_{i,t}$ = Load of consumer $i$ at time $t$
- $G_{j,t}$ = Generation of prosumer $j$ at time $t$
- $L_{i,t}^f$ = Forecasted load of consumer $i$ at time $t$
- $G_{j,t}^f$ = Forecasted generation of prosumer $j$ at time $t$
- $\mathcal{C}_A$ = Set of consumers in Supplier A
- $\mathcal{P}_A$ = Set of prosumers in Supplier A
- $\mathcal{C}_B$ = Set of consumers in Supplier B
- $\mathcal{P}_B$ = Set of prosumers in Supplier B
- $\mathcal{R}$ = Set of all REC members

## Scenario A1: Single Supplier, No REC

**Balancing Group A:**

Actual Net Position at time $t$:

$$\text{Actual Net Position}_t = \sum_{i \in \mathcal{C}_A} L_{i,t} - \sum_{j \in \mathcal{P}_A} G_{j,t}$$

Forecast Net Position at time $t$:

$$\text{Forecast Net Position}_t = \sum_{i \in \mathcal{C}_A} L_{i,t}^f - \sum_{j \in \mathcal{P}_A} G_{j,t}^f$$

Imbalance:

$$\text{Imbalance}_{A,t} = \text{Actual Net Position}_t - \text{Forecast Net Position}_t$$

$$= \left(\sum_{i \in \mathcal{C}_A} L_{i,t} - \sum_{j \in \mathcal{P}_A} G_{j,t}\right) - \left(\sum_{i \in \mathcal{C}_A} L_{i,t}^f - \sum_{j \in \mathcal{P}_A} G_{j,t}^f\right)$$

## Scenario A2: Single Supplier, With REC

**REC Internal Sharing** (at time $t$):

$$\text{Total Load}_t = \sum_{i \in \mathcal{R}} L_{i,t}$$

$$\text{Total Gen}_t = \sum_{j \in \mathcal{R}} G_{j,t}$$

$$\text{Shared Energy}_t = \min(\text{Total Load}_t, \text{Total Gen}_t)$$

$$\text{Exported Energy}_t = \max(0, \text{Total Gen}_t - \text{Total Load}_t)$$

> **Note:** Both Shared Energy and Exported Energy are always non-negative:
> - Shared Energy $= \min(\text{Total Load}_t, \text{Total Gen}_t)$ → Range: $[0, \infty)$
> - Exported Energy $= \max(0, \text{Total Gen}_t - \text{Total Load}_t)$ → Range: $[0, \infty)$

**Corrected Meter Readings:**

For each consumer $i \in \mathcal{R}$:

$$L_{i,t}^{corr} = L_{i,t} - \frac{L_{i,t}}{\text{Total Load}_t} \times \text{Shared Energy}_t$$

For each prosumer $j \in \mathcal{R}$:

$$G_{j,t}^{corr} = \frac{G_{j,t}}{\text{Total Gen}_t} \times \text{Exported Energy}_t$$

This can be rewritten in terms of Shared Energy:

$$G_{j,t}^{corr} = G_{j,t} - \frac{G_{j,t}}{\text{Total Gen}_t} \times \text{Shared Energy}_t$$

### Key Property - Energy Conservation within BG:

**The aggregate net position of a balancing group remains unchanged after REC internal sharing.**

This is the fundamental principle explaining why Scenario A2 has the same imbalance as A1. Even though individual meter readings are corrected to reflect REC sharing, the total load and total generation seen by the balancing group operator are identical before and after the correction.

$$\sum_{i \in \mathcal{C}_A} L_{i,t}^{corr} - \sum_{j \in \mathcal{P}_A} G_{j,t}^{corr} = \sum_{i \in \mathcal{C}_A} L_{i,t} - \sum_{j \in \mathcal{P}_A} G_{j,t}$$

**Proof:**

Starting from the corrected meter readings:

For each consumer $i \in \mathcal{R}$:

$$L_{i,t}^{corr} = L_{i,t} - \frac{L_{i,t}}{\text{Total Load}_t} \times \text{Shared Energy}_t$$

For each prosumer $j \in \mathcal{R}$:

$$G_{j,t}^{corr} = G_{j,t} - \frac{G_{j,t}}{\text{Total Gen}_t} \times \text{Shared Energy}_t$$

Summing over all REC members:

$$\sum_{i \in \mathcal{R}} L_{i,t}^{corr} = \sum_{i \in \mathcal{R}} \left(L_{i,t} - \frac{L_{i,t}}{\text{Total Load}_t} \times \text{Shared Energy}_t\right)$$

$$= \sum_{i \in \mathcal{R}} L_{i,t} - \sum_{i \in \mathcal{R}} \frac{L_{i,t}}{\text{Total Load}_t} \times \text{Shared Energy}_t$$

$$= \sum_{i \in \mathcal{R}} L_{i,t} - \frac{\text{Shared Energy}_t}{\text{Total Load}_t} \sum_{i \in \mathcal{R}} L_{i,t}$$

Since $\sum_{i \in \mathcal{R}} L_{i,t} = \text{Total Load}_t$:

$$\sum_{i \in \mathcal{R}} L_{i,t}^{corr} = \sum_{i \in \mathcal{R}} L_{i,t} - \text{Shared Energy}_t$$

Similarly for generation:

$$\sum_{j \in \mathcal{R}} G_{j,t}^{corr} = \sum_{j \in \mathcal{R}} \left(G_{j,t} - \frac{G_{j,t}}{\text{Total Gen}_t} \times \text{Shared Energy}_t\right)$$

$$= \sum_{j \in \mathcal{R}} G_{j,t} - \sum_{j \in \mathcal{R}} \frac{G_{j,t}}{\text{Total Gen}_t} \times \text{Shared Energy}_t$$

$$= \sum_{j \in \mathcal{R}} G_{j,t} - \frac{\text{Shared Energy}_t}{\text{Total Gen}_t} \sum_{j \in \mathcal{R}} G_{j,t}$$

Since $\sum_{j \in \mathcal{R}} G_{j,t} = \text{Total Gen}_t$:

$$\sum_{j \in \mathcal{R}} G_{j,t}^{corr} = \sum_{j \in \mathcal{R}} G_{j,t} - \text{Shared Energy}_t$$

Therefore:

$$\sum L_{i,t}^{corr} - \sum G_{j,t}^{corr} = (\sum L_{i,t} - \text{Shared Energy}_t) - (\sum G_{j,t} - \text{Shared Energy}_t) = \sum L_{i,t} - \sum G_{j,t}$$

The Shared Energy term cancels out when computing the net position, leaving the aggregate unchanged.

**Imbalance in A2:**

$$\text{Imbalance}_{A,t} = \left(\sum_{i \in \mathcal{C}_A} L_{i,t}^{corr} - \sum_{j \in \mathcal{P}_A} G_{j,t}^{corr}\right) - \left(\sum_{i \in \mathcal{C}_A} L_{i,t}^f - \sum_{j \in \mathcal{P}_A} G_{j,t}^f\right)$$

Since corrected aggregate = original aggregate:

$$\boxed{\text{Imbalance}_{A2} = \text{Imbalance}_{A1}}$$

## Scenario B1: Multiple Suppliers, No REC

**Supplier A (Consumers only):**

$$\text{Actual Net}_A = \sum_{i \in \mathcal{C}_A} L_{i,t} - 0$$

$$\text{Forecast Net}_A = \sum_{i \in \mathcal{C}_A} L_{i,t}^f - 0$$

$$\text{Imbalance}_A = \sum_{i \in \mathcal{C}_A} (L_{i,t} - L_{i,t}^f)$$

**Supplier B (Prosumers only):**

$$\text{Actual Net}_B = 0 - \sum_{j \in \mathcal{P}_B} G_{j,t}$$

$$\text{Forecast Net}_B = 0 - \sum_{j \in \mathcal{P}_B} G_{j,t}^f$$

$$\text{Imbalance}_B = -\sum_{j \in \mathcal{P}_B} (G_{j,t} - G_{j,t}^f)$$

**Total System Imbalance:**

$$\text{Imbalance}_{\text{Total},B1} = \text{Imbalance}_A + \text{Imbalance}_B$$

## Scenario B2: Multiple Suppliers, With REC

**After REC Sharing:**

Supplier A sees:

$$\text{Actual Net}_A = \sum_{i \in \mathcal{C}_A} L_{i,t}^{corr} - 0$$

Supplier B sees:

$$\text{Actual Net}_B = 0 - \sum_{j \in \mathcal{P}_B} G_{j,t}^{corr}$$

**Imbalances:**

$$\text{Imbalance}_A = \sum_{i \in \mathcal{C}_A} L_{i,t}^{corr} - \sum_{i \in \mathcal{C}_A} L_{i,t}^f$$

$$\text{Imbalance}_B = -\sum_{j \in \mathcal{P}_B} G_{j,t}^{corr} + \sum_{j \in \mathcal{P}_B} G_{j,t}^f$$

**Total System Imbalance:**

$$\text{Imbalance}_{\text{Total},B2} = (\sum L_{i,t}^{corr} - \sum L_{i,t}^f) + (-\sum G_{j,t}^{corr} + \sum G_{j,t}^f)$$

### Key Difference:

$$\sum L_{i,t}^{corr} < \sum L_{i,t} \quad \text{(load reduced by sharing)}$$

$$\sum G_{j,t}^{corr} < \sum G_{j,t} \quad \text{(generation reduced by sharing)}$$

Therefore:

$$\boxed{\text{Imbalance}_{\text{Total},B2} \neq \text{Imbalance}_{\text{Total},B1}}$$

## Summary Comparison

| Scenario | BG Structure | REC Effect | Imbalance |
|----------|-------------|------------|-----------|
| A1 | Single BG | No REC | $I_A$ |
| A2 | Single BG | REC internal to BG | $I_A$ (same) |
| B1 | Two BGs | No REC | $I_A + I_B$ |
| B2 | Two BGs | REC crosses BG boundaries | $I'_A + I'_B \neq I_A + I_B$ |

## Mathematical Reason

**Single Supplier (A1 vs A2):**
- REC sharing is **internal redistribution** within one aggregation boundary
- Net position of the balancing group **unchanged**
- Energy shared inside the BG cancels out when calculating aggregate position
- Result: **Same total imbalance**

**Multiple Suppliers (B1 vs B2):**
- REC sharing **crosses aggregation boundaries**
- Each BG's net position **changes independently**
- Supplier A's consumers see reduced load (after sharing)
- Supplier B's prosumers see reduced generation (after sharing)
- The aggregate position of each BG changes
- Result: **Different total imbalance**

### Physical Interpretation

In **Scenario A2**:
- Energy flows within REC members
- But all REC members belong to same Supplier A
- Supplier A still procures same net energy from market
- Imbalance (forecast error) remains unchanged

In **Scenario B2**:
- Energy flows from Supplier B's prosumers to Supplier A's consumers
- Supplier A procures **less** from market (REC sharing reduces their load)
- Supplier B sells **less** to market (REC sharing consumes their generation)
- Both suppliers' market positions change
- Each supplier's imbalance changes relative to their new position
- Total system imbalance is different
